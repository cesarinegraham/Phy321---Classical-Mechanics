### 1i. Coupled Motion and Analytical Motion:
    
We call the motion in the $z$ and $x$ directions as coupled in this case because the motion in the $z$ direction affects the motion in the $x$ direction, and vice versa. This is because the air resistance force in the $x$ direction depends on the velocity in the $x$ direction, which in turn depends on the velocity in the $z$ direction due to the balloon's motion through the air. Similarly, the wind blowing in the $x$ direction affects the motion in the $z$ direction through the air resistance force.

We can determine the motion of the balloon analytically by solving the differential equations for the position and velocity in the $z$ and $x$ directions as functions of time, which we obtained in parts (1e) and (1g) respectively. This will give us the analytical expression for the trajectory of the balloon as it moves through the air, taking into account the effects of buoyancy, air resistance, and wind. However, the resulting equations are quite complicated and may not have a closed-form solution. In general, numerical methods are often used to solve such equations and obtain numerical solutions for the motion of the balloon.

### 1j. Position and Velocity Plots as a Function of Time:

```python
import numpy as np
import matplotlib.pyplot as plt

B = 12 #N
m = 1 #kg
g = 9.81 #m/s/s
D = 1 #unitless constant
w = 1 #m/s

# set up arrays for t, a, v, and z
DeltaT = 0.1
tfinal = 20
n = np.ceil(tfinal/DeltaT)
n = int(n)
t = np.zeros(n)
a = np.zeros(n)
v = np.zeros(n)
z = np.zeros(n)

# Initial conditions
v[0] = 0.0 #m/s
z[0] = 0.0 #m

# Start integrating using Euler’s method
for i in range(n-1):
    # expression for acceleration
    a[i] = (B - m*g - D*abs(v[i]-w)*(v[i]-w)) / m
    # update velocity and position
    z[i+1] = (z[i] + DeltaT*v[i])
    v[i+1] = (v[i] + DeltaT*a[i])
    # update time to next time step and compute analytical answer
    t[i+1] = t[i] + DeltaT

plt.figure(figsize=(16,6))
plt.subplot(1,2,1)
plt.plot(t, z, color='g')
plt.title("position over time")
plt.xlim(0, tfinal)
plt.xlabel('t [s]')
plt.ylabel('z [m]')
plt.grid()

plt.subplot(1,2,2)
plt.plot(t, v, color='g')
plt.title("velocity over time")
plt.xlim(0, 5)
plt.xlabel('t [s]')
plt.ylabel('v [m/s]')
plt.axhline(y=2.479, color='m')
plt.grid()
```

<span style="color:blue">

The plots of position and velocity show that the balloon initially accelerates upward due to the difference between the buoyancy force and the weight of the balloon, and eventually reaches a steady state where the upward buoyancy force is balanced by the downward weight and drag forces. The terminal velocity of the balloon can be estimated by setting the acceleration to zero and solving for the velocity:    
$$
0 = B - m g - D (v_{\text{term}}-w)^2
\\
v_{\text{terminal}} = w + \sqrt{\frac{B-mg}{D}}
\\
\therefore v_{\text{terminal}} = 2.478 \ \frac{m}{s}
$$

### 2a. Forces Acting on the System:

<span style="color:blue">
    
The net force acting on the point mass is given by the vector sum of the gravitational force and the tension force: 
$$
\begin{aligned}
{F}_{net} &= {F}_g+{T} \\
{F}_{net} &= mg + T \\
m\ddot{r} &= mg + T
\end{aligned}
$$

We will now convert the coordinates using the relation $\boldsymbol{r} = l\sin(\phi)\boldsymbol{i}+l\cos(\phi)\boldsymbol{j}$ and take the second derivative with respect to time: 

$$
\begin{aligned}
\ddot{\boldsymbol{r}} &= \frac{d^2}{dt^2}\left(l\sin(\phi)\boldsymbol{i}+l\cos(\phi)\boldsymbol{j}\right) \\
&= l\frac{d^2}{dt^2}\left(\sin(\phi)\boldsymbol{i}+\cos(\phi)\boldsymbol{j}\right) \\
&= l\ddot{\phi}(-\cos(\phi)\boldsymbol{i}+\sin(\phi)\boldsymbol{j})
\end{aligned}
$$

Substituting this expression for $\ddot{\boldsymbol{r}}$ into the equation of motion we get: 

$$
\begin{aligned}
m l \ddot{\phi} \cos(\phi) &= -T \\
m l \ddot{\phi} \sin(\phi) &= -mg - T \\
m l \frac{d^2 \phi}{dt^2} &= -mg\sin(\phi) \\
\ddot{\phi}(t) &= -\omega_0^2\sin{(\phi(t))} \\
\ddot{\phi}(t) &= -\omega_0^2\phi(t)
\end{aligned}
$$

which is the equation of motion for the mathematical pendulum. Note the use of the small angle approximation. 

<span style="color:blue">

### 2b. Analytical Solution of $\ddot{\phi}(t) = -\omega_0^2\phi(t)$:

The general solution to the differential equation $\ddot{\phi}(t)=-\omega_0^2\phi(t)$ is of the form: $\phi(t) = A\cos(\omega_0t) + B\sin(\omega_0t)$, where $A$ and $B$ are constants that can be determined from initial conditions.

$$
\begin{aligned}
\phi(0) &= A\cos(0) + B\sin(0) = A \\
\dot{\phi}(0) &= -A\omega_0\sin(0) + B\omega_0\cos(0) = B\omega_0 \\
A &= \phi_0 \\
B &= \frac{\dot{\phi}_0}{\omega_0}
\end{aligned}
$$

Substituting these values into the general solution, we get:

$$
\phi(t) = \phi_0\cos(\omega_0t) + \frac{\dot{\phi}_0}{\omega_0}\sin(\omega_0t)
$$

which is the analytical solution for the differential equation.

<span style="color:blue">

### 2c. Scaling the Equations:
    
To scale the equations in terms of a dimensionless time $\hat{t}=\omega_0t$, we make the following substitutions:

$$
\begin{aligned}
\frac{d}{dt} &= \frac{d\hat{t}}{dt}\frac{d}{d\hat{t}} = \omega_0\frac{d}{d\hat{t}} \\
\phi &= \hat{\phi} \\
\dot{\phi} &= \frac{d\phi}{dt} = \frac{d\hat{\phi}}{d\hat{t}}\frac{d\hat{t}}{dt} = \omega_0\frac{d\hat{\phi}}{d\hat{t}}
\end{aligned}
$$

Using these substitutions, the two first-order differential equations become:

$$
\begin{aligned}
\frac{d\dot{\hat{\phi}}}{d\hat{t}} &= -\sin{(\hat{\phi})} \\
\frac{d\hat{\phi}}{d\hat{t}} &= \frac{\dot{\hat{\phi}}}{\omega_0} \\
\frac{1}{\omega_0}\frac{d\hat{\phi}}{d\hat{t}} &= \frac{1}{\omega_0}\frac{\dot{\hat{\phi}}}{\omega_0} \\
\frac{d\hat{\phi}}{d\hat{t}} &= \frac{\dot{\hat{\phi}}}{\hat{t}}
\end{aligned}
$$

Now substituting this into the first equation, we get:

$$
\frac{d^2\hat{\phi}}{d\hat{t}^2} = -\sin{(\hat{\phi})}
$$

which is a dimensionless form of the original second-order differential equation in terms of the dimensionless time $\hat{t}=\omega_0t$.
    
### Euler-Cromer Method Algorithm:

1. Define the initial conditions: $\hat{\phi}_0$ and $\dot{\hat{\phi}}_0$.

2. Choose a time step size $\Delta\hat{t}$.

3. Set $n=0$ and $\hat{\phi}_n = \hat{\phi}_0$, $\dot{\hat{\phi}}_n = \dot{\hat{\phi}}_0$.

4. For each time step, $n\rightarrow n+1$, calculate:

   a. $\dot{\hat{\phi}}_{n+1} = \dot{\hat{\phi}}_n - \Delta\hat{t} \cdot \omega_0^2 \sin{(\hat{\phi}_n)}$
   
   b. $\hat{\phi}_{n+1} = \hat{\phi}_n + \Delta\hat{t} \cdot \dot{\hat{\phi}}_{n+1}$
   
5. Repeat step 4 until the desired time is reached.

6. Output the values of $\hat{\phi}_n$ and $\dot{\hat{\phi}}_n$ at each time step.

I am using the Euler-Cromer method because it will conserve energy and I am very familiar with the procedure.

### 2e. Kinetic and Potential Energies:
    
Kinetic Energy in terms of the angular velocity $\dot{\phi}$:

$$
\begin{aligned}
K &= \frac{1}{2}m\left(\frac{d\boldsymbol{r}}{dt}\right)^2 \\ 
K &= \frac{1}{2}m\left(l\dot{\phi}\cos{\phi}\boldsymbol{i}-l\dot{\phi}\sin{\phi}\boldsymbol{j}\right)^2 \\
K &= \frac{1}{2}ml^2\dot{\phi}^2
\end{aligned}
$$

Potential Energy in terms of the height $h$:

$$
\begin{aligned}
V &= mgh \\
V &= mgl(1-\cos{\phi})
\end{aligned}
$$

### 2f. Lagrangian for the Mathematical Pendulum:
    
The Lagrangian $\mathcal{L}$ for the mathematical pendulum with the constraint $r=l$ is given by:

$$
\mathcal{L} = K - V + \lambda(r-l)
$$

where $K$ is the kinetic energy, $V$ is the potential energy, $\lambda$ is the Lagrange multiplier, and $r=l$ is the constraint equation. 

Using the expressions for kinetic and potential energies from 2e, we get:

$$
\mathcal{L} = \frac{1}{2}ml^2\dot{\phi}^2 - mgl(1 - \cos{\phi}) + \lambda(\dot{r}^2 + r^2 - l^2)
$$

where $\dot{r}^2 + r^2 = l^2$, which satisfies the constraint.

Using the Euler-Lagrange equations, we can derive the equations of motion:

$$
\frac{d}{dt}\left(\frac{\partial\mathcal{L}}{\partial\dot{\phi}}\right) - \frac{\partial\mathcal{L}}{\partial\phi} = 0
$$

$$
\frac{d}{dt}\left(\frac{\partial\mathcal{L}}{\partial\dot{r}}\right) - \frac{\partial\mathcal{L}}{\partial r} = 0
$$

We get the following with the first equation, keeping in mind the $r=l$ constraint:

$$
\frac{d}{dt}(ml^2\dot{\phi}) + mgl\sin{\phi} = \lambda(2r\dot{r}) = 0
$$

$$
\frac{d}{dt}(ml^2\dot{\phi}) + mgl\sin{\phi} = 0
$$

$$
\ddot{\phi} + \frac{g}{l}\sin{\phi} = 0
$$

which is the equation of motion for the pendulum with $\omega_0^2= \frac{g}{l}$.

We get the following with the second equation, keeping in mind the $r=l$ constraint::

$$
\frac{d}{dt}(2\lambda r) - mg\sin{\phi} = 0
$$

$$
\lambda = ml\dot{\phi}^2 + mg\cos{\phi}
$$

which is the Lagrange multiplier. Therefore, we have derived the equation of motion $\ddot{\phi}(t)=-\omega_0^2\sin{(\phi(t))}$ with $\omega_0^2=g/l$ and the Lagrange multiplier $\lambda=ml\dot{\phi}^2+mg\cos{(\phi)}$.

Interpretation of $\lambda$: The Lagrange multiplier $\lambda$ is a variable that is introduced to enforce the constraint that the length of the pendulum is constant. It can be thought of as the tension in the string that keeps the mass at a fixed distance from the pivot point. It represents the force required to keep the pendulum moving.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Define constants
g = 9.81  # gravitational acceleration [m/s^2]
l = 1.0   # length of pendulum [m]
omega0 = np.sqrt(g/l)

# Define function for right-hand side of ODE system
def pendulum_ode(t, y):
    """
    Function for right-hand side of ODE system describing
    the motion of a mathematical pendulum.
    """
    phi, dphi = y
    dydt = [dphi, -omega0**2 * np.sin(phi)]
    return dydt

# Define initial conditions
phi0 = np.pi/6  # initial angle [rad]
dphi0 = 0       # initial angular velocity [rad/s]
y0 = [phi0, dphi0]

# Define time array for solution
t_start = 0      # start time [s]
t_end = 10       # end time [s]
num_points = 1000 # number of time points
t_eval = np.linspace(t_start, t_end, num_points)

# Solve ODE system
sol = solve_ivp(pendulum_ode, [t_start, t_end], y0, t_eval=t_eval)

# Extract solution arrays
phi = sol.y[0,:]
dphi = sol.y[1,:]

# Plot results
fig, ax = plt.subplots()
ax.plot(t_eval, phi, label=r"$\phi$")
ax.plot(t_eval, dphi, label=r"$\dot{\phi}$")
ax.set_xlabel("Time [s]")
ax.legend()
plt.grid()
plt.show()
```
